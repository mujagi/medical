```python
import pandas as pd
data = {
   '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
   '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
   '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
   '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)
df
```

```python
# 관객수가 전체평균 관객수 보다 높은 영화를 출력하시오.
total_mean = df['관객 수'].mean()
groupby_mean =  df.groupby('영화')['관객 수'].mean()
groupby_mean[groupby_mean>total_mean]
```

```python
# 2015년 이후 개봉한 영화 데이터에서 영화, 개봉연도 만 출력하시오
filt = df['개봉 연도'] == 2015
df[filt][['영화','개봉 연도']]
```

```python
## (관객수 * 평점)//100 을 적용해서 , 추천 점수 컬럼을 생성하시오. 추천점수 역순 정렬
df['추천 점수'] = (df['관객 수']*df['평점'])//100
df.sort_values('추천 점수',ascending=False)
```

```python
# 개봉연도 역순정렬 출력
df.sort_values('개봉 연도',ascending=False)
```

```python
# 평점 9.20점 이라는 글자를 추가해서 출력하시오
# 1. 타입을 변경해서 적용방법
df['평점'] = df['평점'].astype(str)+'점'
df
```

```python
# 2. 함수를 이용해서 적용하는 방법
import pandas as pd
data = {
   '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
   '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
   '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
   '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)
df
```

```python
def change(grade):
    return str(grade)+'점'
```

```python
df['평점'] = df['평점'].apply(change)
df
```
