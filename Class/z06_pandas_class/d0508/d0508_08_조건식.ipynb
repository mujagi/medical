```python
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
# 조건 키가 170보다 작거나, 200보다 큰 학생을 출력하시오
filt = (df['키']<170)|(df['키']>200)
df[filt]
df.loc[filt,'키']
```

### 데이터 선택 조건

### 키가 평균보다 큰 학생을 출력하시오.

```python
filt = df['키']>df['키'].mean()
df.loc[filt,['이름','키']]
```

### 국어 85점 이상, 영어 85점이상 이름,학교,국어,영어 출력하시오

```python
filt = (df['국어']>=85) & (df['영어']>=85)
df.loc[filt,['이름','학교','국어','영어']]
df.loc[~filt,['이름','학교','국어','영어']] # ~ : 조건의 반대
```
