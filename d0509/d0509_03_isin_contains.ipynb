```python
import pandas as pd

df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
## 2개 이상 조건 검색부분
lang = ['Python','Java']
filt = df['sw특기'].isin(lang)
df[filt]
```

```python
# startswith, contains, lower, upper
# 문자열 함수 적용 - lower()
lang = ['python','java']
# 2개 이상의 검색을 할 수 있도록 함, isin 명령어 사용
filt = df['sw특기'].str.lower().isin(lang)
df[filt]
```

```python
# java라는 글자가 포함되어있는 것을 검색
# filt = df['sw특기'].str.lower().contains('java',na=False) # contains 는 리스트안에 non이 있을 경우 오류 남 na=False 또는 True로 변경 후 실행필요
filt =  df['sw특기'].str.lower()
filt2 = filt.str.contains('java',na=False)
df[filt2]
```
