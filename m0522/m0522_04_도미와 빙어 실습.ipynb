```python
# 로지스틱 회귀를 적용해서 
# 도미와 빙어 test 세트의 5개를 분류하고 
# 5개는 몇 % 확률로 선택을 했는지 출력하시오

# train,test 정답률을 출력하시오
```

```python
length = np.array(bream_length + smelt_length)
weight = np.array(bream_weight + smelt_weight)
data = np.column_stack((length,weight))
target = np.array(['도미']*35+['빙어']*14)
target
```

```python
data = [target,length,weight]
df = pd.DataFrame(data)
df = df.T
df.head(5)
```

```python
fish_input = df.iloc[:,1:].to_numpy()
fish_target = df[0].to_numpy()

train_input,test_input,train_target,test_target = train_test_split(
    fish_input,fish_target,random_state=42
)
```

```python
# 표준화 작업
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
# 데이터 실습시키기
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(train_scaled,train_target)
```

```python
# 데이터 예측
predict = lr.predict(test_scaled)
print(predict)
```

```python
print(lr.score(train_scaled,train_target))
print(lr.score(test_scaled,test_target))
```
