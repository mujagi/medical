```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows 한글 폰트
matplotlib.rcParams['font.size'] = '15' # font size
matplotlib.rcParams['axes.unicode_minus'] = False # - 기호 깨짐 현상을 해결함

x = [1,2,3]
y = [2,4,8]
# 그래프 크기 키우기
plt.figure(dpi=200)


plt.plot(x,y,label = '국어',ls='-.',marker='o')
plt.legend(loc='upper center')
# plt.savefig('g.png',dpi=300)

plt.show
```

```python
plt.figure(dpi =100)
plt.plot(x,y,label = '국어',ls='-.',marker='o')
plt.legend(loc='upper center')
# 좌표에 데이터 2,4,8의 값을 출력
# plt.text()
# x = [1,2,3] y = [2,4,8] 데이터값
for i,txt in enumerate(y):  
    plt.text(x[i],y[i]+0.2,txt,ha='center',color='blue')
    
plt.yticks([x for x in range(10)])

plt.show
```
