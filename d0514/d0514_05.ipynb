```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows  
matplotlib.rcParams['font.size'] = '15' # font size
matplotlib.rcParams['axes.unicode_minus'] = False # -    

x = [1,2,3]
y = [2,4,8]
#   
plt.figure(dpi=200)


plt.plot(x,y,label = '',ls='-.',marker='o')
plt.legend(loc='upper center')
# plt.savefig('g.png',dpi=300)

plt.show
```

```python
plt.figure(dpi =100)
plt.plot(x,y,label = '',ls='-.',marker='o')
plt.legend(loc='upper center')
#   2,4,8  
# plt.text()
# x = [1,2,3] y = [2,4,8] 
for i,txt in enumerate(y):  
    plt.text(x[i],y[i]+0.2,txt,ha='center',color='blue')
    
plt.yticks([x for x in range(10)])

plt.show
```
