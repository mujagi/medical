```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows 한글 폰트
matplotlib.rcParams['font.size'] = '15' # font size
matplotlib.rcParams['axes.unicode_minus'] = False # - 기호 깨짐 현상을 해결함

x = [1,2,3]
y = [12,9,8]

# plt.plot(x,y,linewidth=3,marker='*',markersize=10,markeredgecolor='red',label = '국어') # linewidth 선두께, laber 범례 
# markersize -> ms 마커크기
# markedgecolor -> mec 마커둘레색상
# markerfacecolor -> mfc 마커색상
# linestyle -> ls 선표현
plt.plot(x,y,linewidth=3,marker='*',ms=10,mec='red',mfc='yellow',alpha=0.7,ls=':',label = '국어') # linewidth 선두께, laber 범례 
plt.legend(loc=(0.5,0.5)) # 위치지점 변경
plt.title('막대그래프')
```
