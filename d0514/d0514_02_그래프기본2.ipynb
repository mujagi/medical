```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows 한글 폰트
matplotlib.rcParams['font.size'] = '15' # font size
matplotlib.rcParams['axes.unicode_minus'] = False # - 기호 깨짐 현상을 해결함
```

```python
x= ['han','kim','hong']
y= [20,40,80]

plt.plot(x,y)
plt.title('막대그래프',fontdict={'family':'D2CODING','size':30}) # title 글꼴, 크기 변경
# x,y 축 라벨 설정
plt.xlabel('이름',loc='left',color='red') # left, right, center(default)
plt.ylabel('국어',loc='top',color='green') # top, bottom, center(default)
```

```python
x= ['han','kim','hong']
y= [20,40,80]
y2= [50,30,90]
y3= [80,90,95]

plt.plot(x,y,label='국어') # label = : 범례명 지정
plt.plot(x,y2,label='영어') # label = : 범례명 지정
plt.plot(x,y3,label='수학') # label = : 범례명 지정
plt.legend(loc=4,ncol=3,prop={'size':10}) # 범례 표시 (1:우상단, 2:좌상단, 3:좌하단, 4:우하단) // ncol = : 범례를 가로 표시
plt.title('막대그래프',fontdict={'family':'D2CODING','size':30}) # title 글꼴, 크기 변경
# x,y 축 라벨 설정
plt.xlabel('이름',loc='left',color='red') # left, right, center(default)
plt.ylabel('국어',loc='top',color='green') # top, bottom, center(default)
```
