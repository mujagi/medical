```python
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows 한글 폰트
matplotlib.rcParams['font.size'] = '10' # font size
matplotlib.rcParams['axes.unicode_minus'] = False # - 기호 깨짐 현상을 해결함
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
# 이름과 국어 성적을 가지고 선 그래프를 출력하시오
x = df['이름']
y = df['국어']
plt.title('빅데이터 성적그래프')
plt.bar(x,y,label='국어점수')
plt.legend(loc='upper center')
```

```python
# chipotle.tsv에서 주문중 가장 많이 주문한 메뉴 10개를 꺽은선 그래프로 출력하시오
import pandas as pd
df = pd.read_csv('../z20_data/chipotle.tsv',sep='\t')
df.fillna('',inplace=True)
x = df.groupby('item_name')['quantity'].sum().sort_values(ascending=False).head(5).index
y = df.groupby('item_name')['quantity'].sum().sort_values(ascending=False).head(5).values
plt.plot(x,y)
```

```python
# item_name 별 주문한 개수의 합 출력하시오.
# item_name별 주문한 횟수가 몇번인지 출력하시오.
y2 = df['item_name'].value_counts()[:5]
plt.plot(x,y,label = '주문한 개수')
plt.plot(x,y2,label = '주문횟수')
plt.legend(loc='upper right')
```
