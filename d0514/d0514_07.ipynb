```python
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows 한글 폰트
matplotlib.rcParams['font.size'] = '10' # font size
matplotlib.rcParams['axes.unicode_minus'] = False # - 기호 깨짐 현상을 해결함
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
x = df.head(8)['이름']
y = df.head(8)['국어']
y2 = df.head(8)['영어']
y3 = df.head(8)['수학']
```

```python
plt.plot(x,y,label='국어',marker='D')
plt.plot(x,y2,label='영어',marker='o',ls='--')
plt.plot(x,y3,label='수학',marker='s',ls='-.')

for i,txt in enumerate(y3) :
    plt.text(x[i],y3[i]+0.9,txt,ha='center')

plt.legend(loc='upper center',ncol=3) #ncol 범례 1줄에 3개까지
plt.show
```

```python
colors = ['r','g','b','y','pink']
plt.bar(x,y,label='국어',color=colors)

for i,txt in enumerate(y) :
    plt.text(x[i],y[i]+0.9,txt,ha='center')

plt.xticks(rotation=45)

plt.legend(loc='upper center',ncol=3) #ncol 범례 1줄에 3개까지
plt.show
```
