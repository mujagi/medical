```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
import seaborn as sns
matplotlib.rcParams['axes.unicode_minus'] = False # - 기호 깨짐 현상을 해결함
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows 한글 폰트
matplotlib.rcParams['font.size'] = '10' # font size
```

```python
df = pd.read_csv('../z20_data/diabetes.csv')
df.columns
```

### 나이가 50이상인 사람은 몇명인지?

```python
filt = df['Age'] >= 50
len(df[filt])
```

```python
# 나이가 50이상이면서 당뇨병 환자인사람은 몇명인지 출력하시오
filt = (df['Age']>=50) & (df['Outcome']==1)
len(df[filt])
```

```python
# 각 항목별 상관관계 함수
data_corr = df.corr()
fig = plt.figure(figsize=(7,7))
sns.heatmap(data_corr,annot=True)
plt.show()
```

```python
explode = [0,0.1]
values = df['Outcome'].value_counts()
labels =['안환자','환자']
plt.pie(values,labels=labels,autopct='%.1f%%',startangle=90,explode=explode)
plt.title('당뇨병 환자분포') 
```
