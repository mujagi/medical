### 연령별 인구현황
 - 2023년 1월 연령별 인구수를 가로 막대 그래프로 출력하시오.

```python
m1 = pd.read_excel('../z20_data/20230104_연령별인구현황.xlsx',skiprows=3,index_col='행정기관',usecols='B,E:Y')
fm1 = pd.read_excel('../z20_data/20230104_연령별인구현황.xlsx',skiprows=3,index_col='행정기관',usecols='B,AB:AV')

m1.iloc[0] = m1.iloc[0].str.replace(',','').astype(int)
fm1.iloc[0] = fm1.iloc[0].str.replace(',','').astype(int)
fm1.columns = fm1.columns.str.replace('.1','')
    
```

```python

x = m1.columns
y = fm1.iloc[0,0:]//1000
plt.barh(x,y)
x = fm1.columns
y1 = fm1.iloc[0,0:]//1000
plt.barh(x,-y1)


plt.show
```

```python
# 1월,2월,3월,4월 합산을 해서 가로막대 그래프를 출력하시오
df = pd.read_excel('../z20_data/20230104_연령별인구현황.xlsx',skiprows=3,thousands = ',',index_col='행정기관')
```

```python
df.iloc[0,3:24] #남자 1월
df.iloc[0,26:47] # 여자 1월
df.iloc[0,49:70] # 남자 2월
df.iloc[0,72:93] # 여자 2월
df.iloc[0,95:116] # 남자 3월
df.iloc[0,118:139] # 여자 3월
df.iloc[0,141:162] # 남자 4월
df.iloc[0,164:185] # 여자 4월

man = (df.iloc[0,3:24].values +df.iloc[0,49:70].values + df.iloc[0,95:116].values+df.iloc[0,141:162].values)//1000
girl = (df.iloc[0,26:47].values +df.iloc[0,72:93].values + df.iloc[0,118:139].values+df.iloc[0,164:185].values)//1000
x = m1.columns
y = man
y1 = girl
plt.barh(x,y)
plt.barh(x,-y1)
```
