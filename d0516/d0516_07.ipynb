```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows 한글 폰트
matplotlib.rcParams['font.size'] = '10' # font size
matplotlib.rcParams['axes.unicode_minus'] = False # - 기호 깨짐 현상을 해결함
```

```python
df = pd.read_excel('../z20_data/stat_142801.xls',skiprows=2,nrows=2,index_col=0)
df
```

```python
# 엑셀파일을 import 할 때 index에 보이지 않는 특수문자가 있는지 확인
df.index.values
```

```python
df.rename(index = {'출생아\xa0수':'출생아 수','합계\xa0출산율':'합계 출산율'},inplace=True)
```

```python
df.loc['출생아 수']
```

```python
df.loc['합계 출산율']
```

```python
plt.bar(df.columns,df.loc['출생아 수'])
```

```python
# 출생아 수는 막대그래프
# 합계 출산율 꺽은선 그래프로
plt.bar(df.columns,df.loc['출생아 수'],color='orange')

for i,txt in enumerate(df.loc['출생아 수']) :
    plt.text(df.columns[i],df.loc['출생아 수'][i]+0.5,txt,ha='center')

plt.ylim(200,700)
plt.yticks([300,400,500,600])
```

```python
plt.plot(df.columns,df.loc['합계 출산율'],color='yellow',marker='o')
for i,txt in enumerate(df.loc['합계 출산율']) :
    plt.text(df.columns[i],df.loc['합계 출산율'][i],txt,ha='center')
plt.ylim(0,1.5)
plt.yticks([0,1])
```

```python
fig,ax1 = plt.subplots(figsize = [10,7])
ax1.bar(df.columns,df.loc['출생아 수'],color = '#ff812d')
ax1.set_ylim(200,700)
ax1.set_yticks([300,400,500,600])
for i,txt in enumerate(df.loc['출생아 수']) :
    plt.text(df.columns[i],df.loc['출생아 수'][i]+3,txt,ha='center')
ax1.set_ylabel('출생아 수')

# 1개의 그래프에 두개 넣기
ax2=ax1.twinx()

ax2.plot(df.columns,df.loc['합계 출산율'],color = '#ffd100',marker='o',markersize=10,markeredgecolor='w',markeredgewidth=3,linewidth=3)
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1])
for i,txt in enumerate(df.loc['합계 출산율']) :
    plt.text(df.columns[i],df.loc['합계 출산율'][i]+0.05,txt,ha='center')
```
